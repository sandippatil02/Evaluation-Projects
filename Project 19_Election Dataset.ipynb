```python
# Load libraries
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sb
```

```python
# Load dataset
ds = pd.read_csv('ElectionData.csv')
print(ds.isnull().any())
```

```python
# Let's get Unique values from a column
k = ds.nunique()
```

```python
# Drop the TimeElapsed & the Time columns
ds = ds.drop(['TimeElapsed', 'time'], axis = 1)
```

```python
# Shuffle the dataframe
ds = ds.sample(frac = 1).reset_index(drop = True)
```

```python
X = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values
```

```python
# Let's Encode the territory column
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
lbl = LabelEncoder()
```

```python
X[:, 0] = lbl.fit_transform(X[:, 0])
X[:, -6] = lbl.fit_transform(X[:, -6])
```

```python
ct = ColumnTransformer(transformers = [('one_hot_encoder', OneHotEncoder(categories = 'auto'), [0, -6])],
                       remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X = X[:, 1:]
```

```python
# Splitting into training & testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Fitting the regressor to the training data
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train, y_train)
```

```python
# Predictions
y_pred = regressor.predict(X_test)
```

```python
# Error
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
r2_score = r2_score(y_test, y_pred)
explained_variance_score = explained_variance_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
print("\nMean Squared Error: ", mse)
print("\nR2 Score: ", r2_score)
print("\nExplained Variance Score: ", explained_variance_score)
```

```python
print("\nExplained Variance Score:",explained_variance_score)
```

```python
print("\nMean Squared Error:",mse)
```

```python
print("\nR2 Score:", r2_score)
```
