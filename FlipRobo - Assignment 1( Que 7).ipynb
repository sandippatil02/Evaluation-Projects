```python
# Import essential libraries

import numpy as np
import pandas as pd

from selenium import webdriver
from bs4 import BeautifulSoup

import requests
import urllib.request
from urllib.request import urlopen
```

```python
# Configure webdriver to use Chrome Browser

driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.amazon.in/s/?k=phones+under+20000+with+8gb+ram&ref=sugsr_0&pd_rd_w=kSMnN&pf_rd_p=e98d1fad-4664-4b4d-a930-8932601ebadf&pf_rd_r=A7KBM8W44NZMHY6PNGHX&pd_rd_r=ae3f81d4-481f-487f-a9a0-2c837e730317&pd_rd_wg=cvo3c&qid=1604055312")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup)
```

```python
name = soup.find_all('span', class_ = 'a-size-medium a-color-base a-text-normal')
name[0].text
```

```python
Product_Name = []
for i in name:
    Product_Name.append(i.text)

    
len(Product_Name)
```

```python
price = soup.find_all('span', class_ = 'a-price-whole')
Price = []
for i in price:
    Price.append(i.text)
    
len(Price)
```

```python
rating = soup.find_all('span', class_ = 'a-icon-alt')

Average_Rating = []

for i in rating:
    Average_Rating.append(i.text)
    
Average_Rating
```

```python
unwanted_element= ['4 Stars & Up','3 Stars & Up','2 Stars & Up','1 Star & Up', ' out of 5 stars']
Average_Ratings=[i for i in Average_Rating if i not in unwanted_element]
```

```python
Average_Ratings
```

```python
image = soup.find_all('div', class_ = 'a-section aok-relative s-image-fixed-height')

Image_URL = []

for i in range(len(image)):
    Image_URL.append(image[i].img['src'])
```

```python
image[0].img['src']
```

```python
Image_URL
```

```python
print(len(Product_Name))
print(len(Price))
print(len(Average_Ratings))
print(len(Image_URL))
```

```python
df=pd.DataFrame({"Product_Name":Product_Name,"Price":Price, "Average_Ratings":Average_Ratings, "Image_URL": Image_URL})
df.to_csv('Mobile_phones_under_Rs_20,000.csv', index = False, encoding = 'utf-8-sig')
```
