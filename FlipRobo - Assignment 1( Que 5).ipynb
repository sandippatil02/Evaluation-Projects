 ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
Player=[]
Team=[]
Rating=[]
driver.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
```

```python
content= driver.page_source
soup= BeautifulSoup(content)
player=soup.findAll("td", attrs={"class":"table-body__cell rankings-table__name name"})
team=soup.findAll("span", attrs={"class":"table-body__logo-text"})
rating=soup.findAll("td", attrs={"class":"table-body__cell rating"})
print(len(player), len(team), len(rating))
```

```python
for i in player:
    Player.append(i.text)
```

```python
Player
```

```python
NewPlayer=[]
for i in Player:
    newname=i.replace("\n","")
    NewPlayer.append(newname)
```

```python
NewPlayer
```

```python
for i in team:
    Team.append(i.text)
```

```python
Team
```

```python
for i in rating:
    Rating.append(i.text)
```

```python
Rating
```

```python
df=pd.DataFrame({"PlayerName":NewPlayer,"TeamName":Team, "Ratings":Rating})
```

```python
df.head()
```

```python
df.iloc[:10,:]
```
